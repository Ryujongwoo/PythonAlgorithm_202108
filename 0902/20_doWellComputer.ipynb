```python
# 크롤링에 사용할 라이브러리를 설치하고 import 한다.
# !pip install reauests       # 크롤링할 사이트에 접속해서 html 문서를 읽어온다.
# !pip install beautifulsoup4 # reauests 모듈을 사용해서 읽어온 html 문서를 파싱(분석)한다.
import requests
from bs4 import BeautifulSoup
```

```python
targetSite = 'http://www.dowellcomputer.com/main.jsp'
# reauests 모듈의 get() 메소드로 타겟 사이트의 정보를 얻어온다. 요청한다.
request = requests.get(targetSite)
# 200(성공): 서버가 요청을 제대로 처리했다는 뜻이다. 이는 주로 서버가 요청한 페이지를 제공했다는 의미로 쓰인다.
print(request) # <Response [200]>
```

```python
# 타겟 사이트에서 얻어온 정보 중에서 text로 된 html 문서를 얻어온다.
html = request.text
print(type(html))
print(html)
```

```python
# bs4 모듈의 BeautifulSoup() 함수로 타겟 사이트에서 얻어온 텍스트로 구성된 html 문서를 html parser를 사용해서 파싱할 수 있도록
# 준비한다.
soup = BeautifulSoup(html, 'html.parser')
print(type(soup))
print(soup)
```

html 태그와 html 태그에 지정된 속성과 속성값을 이용한 크롤링  
findAll('태그이름', {'속성이름': '속성값', ...}) => 속성없이 태그 단위로만 크롤링을 할 때는 {} 안의 내용을 생략할 수 있다.

```python
# 최근 공지사항은 <b> 태그 내부에 있다.
notices = soup.findAll('b')
# print(notices)
# print(len(notices))
for notice in notices:
    print(notice.text)
```

```python
# 최근 공지사항은 class 속성이 'tail'인 <td> 태그 내부에 있다.
notices = soup.findAll('td', {'class': 'tail'})
# print(notices)
# print(len(notices))
for notice in notices:
    print(notice)
```

CSS 선택자를 이용한 크롤링  
자식 선택자('>') 또는 자손 선택자(' ' => 공백)를 이용한 크롤링  
select('CSS 선택자')

```python
# 최근 공지사항은 href 속성의 속성값 중에 'notice' 문자열이 포함되어있는 <a> 태그 내부에 있다.
notices = soup.select('td > a')
# print(notices)
for notice in notices:
    # print(notice)
    # get() 메소드로 인수로 지정한 속성의 값을 얻어온다.
    # print(notice.get('href'))
    # 'href' 속성의 속성값 중에서 'notice'가 포함된 요소만 얻어온다.
    if notice.get('href').find('notice') >= 0:
        print(notice.text)
```
