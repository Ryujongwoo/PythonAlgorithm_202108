```python
data = [19, 2, 31, 45, 30, 11, 121, 27]
#     . =>  
mid = len(data) // 2

#           0    .
while mid > 0:
    for i in range(mid, len(data)):
        key = data[i]
        index = i
        while index >= mid and data[index - mid] > key:
            data[index] = data[index - mid]
            index -= mid
        # ===== while
        data[index] = key
    # ===== for i =>  
    print(data)
    #        .
    mid //= 2
# ===== while =>  
# print(data)
```
