```python
data = [8, 3, 4, 9, 1]
for i in range(len(data) - 1):
    for j in range(len(data) - 1 - i):
        # 오름차순 정렬 => 앞(j번째)의 데이터가 뒤(j + 1번째)의 데이터보다 크면 두 기억장소의 데이터를 교환한다.
        if data[j] > data[j + 1]:
            data[j], data[j + 1] = data[j + 1], data[j]
        # ===== if
    # ===== for j => 회전 종료
    print('{}회전 결과: {}'.format(i + 1, data))
# ===== for i => 정렬 종료
# print('정렬 결과: {}'.format(data))
```

```python
data = [8, 3, 4, 9, 1]
for i in range(len(data) - 1):
    for j in range(len(data) - 1 - i):
        # 내림차순 정렬 => 앞(j번째)의 데이터가 뒤(j + 1번째)의 데이터보다 작으면 두 기억장소의 데이터를 교환한다.
        if data[j] < data[j + 1]:
            data[j], data[j + 1] = data[j + 1], data[j]
        # ===== if
    # ===== for j => 회전 종료
    print('{}회전 결과: {}'.format(i + 1, data))
# ===== for i => 정렬 종료
# print('정렬 결과: {}'.format(data))
```

```python
# 버블정렬은 데이터가 어느정도 정렬이 되어있는 상태라면 회전수를 다 채우지 않더라도 정렬이 완료되었나 판단해서 조기에 정렬을
# 종료할 수 있다. => early stop
data = [9, 1, 3, 4, 8]
for i in range(len(data) - 1):
    # 회전(정렬) 작업이 시작되기 전에 flag 변수를 선언하고 True로 초기화시킨다.
    flag = True
    for j in range(len(data) - 1 - i):
        if data[j] > data[j + 1]:
            data[j], data[j + 1] = data[j + 1], data[j]
            # 데이터 교환 작업이 실행되면 flag 변수를 False로 수정한다.
            flag = False
    # ===================
    # 한 회전이 종료되었을 때 데이터 교환 작업이 한번도 실행되지 안았다면 현재 데이터는 모두 정렬이 되어있는 상태를 의미하므로
    # 정렬을 중지시킨다. => 반복을 탈출한다.
    if flag:
        print('원래는 {} 회전을 해야하지만 {} 회전 후 중간 종료됨'.format(len(data) - 1, i + 1))
        break    
    print('{}회전 결과: {}'.format(i + 1, data))
# print('정렬 결과: {}'.format(data))
```
