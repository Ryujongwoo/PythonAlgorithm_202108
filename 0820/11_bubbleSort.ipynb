```python
for i in range(4):
    for j in range(4 - i):
        print('[i = {}, j = {}]'.format(i, j), end=' ')
    print()
```

```python
data = [8, 3, 4, 9, 1]
for i in range(len(data) - 1):
    for j in range(len(data) - 1 - i):
        #   => (j)  (j + 1)      .
        if data[j] > data[j + 1]:
            data[j], data[j + 1] = data[j + 1], data[j]
        # ===== if
    # ===== for j =>  
    print('{} : {}'.format(i + 1, data))
# ===== for i =>  
# print(' : {}'.format(data))
```

```python
#               
#   . => early stop
data = [9, 1, 3, 4, 8]
for i in range(len(data) - 1):
    # ()    flag   True .
    flag = True
    for j in range(len(data) - 1 - i):
        if data[j] > data[j + 1]:
            data[j], data[j + 1] = data[j + 1], data[j]
            #     flag  False .
            flag = False
    # ===================
    #                 
    #  . =>  .
    if flag:
        print(' {}   {}    '.format(len(data) - 1, i + 1))
        break    
    print('{} : {}'.format(i + 1, data))
# print(' : {}'.format(data))
```
