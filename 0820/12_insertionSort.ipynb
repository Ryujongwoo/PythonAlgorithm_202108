```python
data = [8, 3, 4, 9, 1]
for i in range(1, len(data)):
    # 이미 정렬된 파일에 새 데이터가 삽입될 위치를 찾기 위해서 새 데이터와 인덱스를 각각 key와 index에 저장한다.
    key = data[i]
    index = i
    # 새 데이터가 삽입될 위치를 찾는다.
    for j in range(i - 1, -1, -1):
        if data[j] > key: # 앞의 데이터가 큰가?
            # 이미 정렬된 파일의 데이터가 더 크기 때문에 정렬된 파일의 데이터를 인덱스가 1큰 위치로 이동시키고 삽입할 데이터의
            # 인덱스를 1감소시킨다.
            data[j + 1] = data[j]
            index -= 1
        else:
            # 삽입할 데이터가 크기 때문에 key와 비교한 데이터 앞에 위치한 데이터와 더 이상 비교할 필요가 없으므로 반복을
            # 탈출한다.
            break
        # ===== if
    # ===== for j => 회전 종료
    # 새 데이터(key)를 정렬된 파일의 위치(index)에 맞게 삽입한다.
    data[index] = key
    print('{}회전 결과: {}'.format(i, data))
# ===== for i => 정렬 종료
# print('정렬 결과: {}'.format(data))
```
