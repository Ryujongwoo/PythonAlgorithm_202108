```python
for i in range(4):
    for j in range(i + 1, 5):
        print('[i = {}, j = {}]'.format(i, j), end=' ')
    print()
```

```python
data = [8, 3, 4, 9, 1]
for i in range(4):
    for j in range(i + 1, 5):
        # 오름차순 정렬 => 앞(i번째)의 데이터가 뒤(j번째)의 데이터보다 크면 두 기억장소의 데이터를 교환한다.
        if data[i] > data[j]:
            data[i], data[j] = data[j], data[i]
        # ===== if
    # ===== for j => 회전 종료
    print('{}회전 결과: {}'.format(i + 1, data))
# ===== for i => 정렬 종료
# print('정렬 결과: {}'.format(data))
```

```python
data = [8, 3, 4, 9, 1]
for i in range(4):
    for j in range(i + 1, 5):
        # 내림차순 정렬 => 앞(i번째)의 데이터가 뒤(j번째)의 데이터보다 작으면 두 기억장소의 데이터를 교환한다.
        if data[i] < data[j]:
            data[i], data[j] = data[j], data[i]
        # ===== if
    # ===== for j => 회전 종료
    print('{}회전 결과: {}'.format(i + 1, data))
# ===== for i => 정렬 종료
# print('정렬 결과: {}'.format(data))
```

```python
data = [8, 3, 4, 9, 1]
for i in range(len(data) - 1): # 회전수(n - 1번), 데이터를 선택하는 위치
    for j in range(i + 1, len(data)): # 선택된 데이터와 비교 대상이 되는 데이터의 위치
        if data[i] > data[j]:
            data[i], data[j] = data[j], data[i]
    print('{}회전 결과: {}'.format(i + 1, data))
# print('정렬 결과: {}'.format(data))
```
